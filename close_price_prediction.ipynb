```python
import pandas as pd
#import datetime
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn import metrics
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('ohlc.csv')
df.head
```

```python
df.drop(['Volume USD'], axis=1, inplace=True)
```

```python
df.describe()
```

```python
df.info()
```

```python
df.drop(['unix'], axis=1, inplace=True)
```

```python
df.drop(['symbol'], axis=1, inplace=True)
```

```python
df.rename(columns={'Volume BTC':'volume'}, inplace=True)
```

```python
date = df['date'].str.replace(' ', '-').str.split('-', expand=True)
year = date[0].astype('int')
month = date[1].astype('int')
day = date[2].astype('int')
df['year'] = year
df['month'] = month
df['day'] = day
df
```

```python
df.isnull().sum()
```

```python
data_grouped = df.groupby('year').mean()
plt.subplots(figsize=(20,10))
for i, col in enumerate(['open', 'high', 'low', 'close']):
  plt.subplot(2,2,i+1)
  data_grouped[col].plot.bar()
plt.show()
```

```python
plt.figure(figsize=(5,5))
sb.heatmap(df.corr(), annot=True, cbar=False)
plt.show()
```

```python
df['target'] = np.where(df['close'].shift(-1) > df['close'], 1, 0)
df
```

```python
df = df.loc[::-1]
```

```python
plt.pie(df['target'].value_counts().values, labels=[0,1], autopct="%.2f%%")
```

```python
df
```

```python
features = df[['open', 'high', 'low', 'volume', 'month', 'day', 'year']]
target = df['close']

scaler = StandardScaler()
features = scaler.fit_transform(features)

X_train, X_test, Y_train, Y_test = train_test_split(features, target, test_size=0.1, random_state=2022)
print(X_train.shape, X_test.shape)
```

```python
regr = LinearRegression()
regr.fit(X_train, Y_train)
```

```python
x_features = np.asanyarray(df[['open', 'high', 'low', 'volume', 'month', 'day', 'year']])
y_target = np.asanyarray(df['close'])

predicted_output = regr.predict(X_test)
print("R-squared score: %.2f" % r2_score(Y_test, regr.predict(X_test)))
print("Mean absolute error: %.2f" % np.mean(np.absolute(predicted_output - Y_test)))
print("Residual sum of squares (MSE): %.2f" % np.mean((predicted_output - Y_test) ** 2))
print('Variance score: %.2f' % metrics.explained_variance_score(Y_test, regr.predict(X_test)))
```
